```python
def CTRs_plot(CTR,var,t,k,iterations):
    var = str(var)
    t = str(t)
    k = str(k)
    iterations = str(iterations)
    ctrs = [str(i) for i in range(0,len(CTR))]
    index = np.arange(len(ctrs))
    bar_width = 0.6
    fig, ax = plt.subplots()
    ax.bar(index, CTR, bar_width,  color="darkblue")
    ax.set_yticks([0,0.1,0.2,0.3,0.4,0.6,0.6,0.7,0.8,0.9,1])
    ax.set_ylabel("CTR")
    ax.set_xlabel("Advertiser ID")
    ax.set_title("CTRs UCB1 plot with user attractiveness variance:"+var+" t:"+t+" k:"+k+" iterations:"+iterations)
```
