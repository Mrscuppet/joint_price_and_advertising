```python
import random
from collections import namedtuple
import numpy as np
import copy
import JPBenvironment as envv
import JPBEnvCTR as envctr
import JPBagents as age
import utility as ut
import matplotlib.pyplot as plt
import numpy as np
```

```python
comp = namedtuple('Agent', ['id', 'price','bid'])
comp1 = namedtuple('Competitor', ['id', 'price','bid'])
```

```python
#Lista coppie bi e price competitor
Shop0 = comp1(0,7.8,2.5)
Shop1 = comp1(1,6.5,2.4)
Shop2 = comp1(2,7,2.44)
Shop3= comp1(3,7.8,2.75)
Shop4 = comp1(4,7.16,2.46)

#Agent, per ora considero solo una coppia per vedere se riesco a farlo partire
#Shop5 = comp(5,5.2,2.2)

agente = []

ag_0 = comp(5,7.2,2.43)
ag_1 = comp(6,6.2,2.3)
ag_2 = comp(7,8.05,2.7)
ag_3 = comp(8,6.99,2.35)
ag_4 = comp(9,5.1,1.48)

agente.append(ag_0)
agente.append(ag_1)
agente.append(ag_2)
agente.append(ag_3)
agente.append(ag_4)

shops = []

shops.append(Shop0)
shops.append(Shop1)
shops.append(Shop2)
shops.append(Shop3)
shops.append(Shop4)
#shops.append(Shop5)

shops
```

```python
lista_CTRs = []
```

```python
#Inizializzo CTR
CTR = [1,1,1,1,1,1]
```

```python
#Environment
env1 = envv.Environment(2,10,ut.allocation_mec,ut.user_gernator,CTR)

#MAB
maxrewards  = [agente[i].price - agente[i].bid for i in range(len(agente))]
mab0 = age.UCB1Agent(len(agente),maxrewards)

for _ in range(100):
    index_agent_pulled_arm = mab0.pull_arm()
    reward = env1.step(shops[:5],agente[index_agent_pulled_arm])
    mab0.update(reward[0])

```
